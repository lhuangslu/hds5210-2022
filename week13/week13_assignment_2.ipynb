```python
# Each state has different number of children.  To compare total victims by state, it is better to use rates (i.g. per 1000 children). 
# I summed up the columns that report the rates, then plotted in the histogram.  

col_rate=[]
col_position=[]

for k, v in enumerate(data.columns):
    if "Rate" in v:
        col_rate.append(v)
        col_position.append(k)
        
col_position
```

```python
col_rate
```

```python
data['Total Per 1000']= data.iloc[:, col_position[0]:].sum(axis=1)
```

```python
import matplotlib.pyplot as plt

data['Total Per 1000'].hist()  # It looks like the distribution is bimodal 
plt.xlabel('Rate of Total Child Victims (per 1000 childen)')
plt.title('Fig. Child Abuse of 52 States')
```

```python
data[data["Total Per 1000"]==data['Total Per 1000'].max()]["State"]    # Kentucky reported the highest rate of child abuse.  
```
